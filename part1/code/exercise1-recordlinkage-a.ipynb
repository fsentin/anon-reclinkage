```python
f1['age'] = f1['age'].replace(['[0, 40[', '[40, 90['], '[0, 90[')
```

```python
f1 = f1.drop(columns=['workclass', 'education', 'marital-status', 'occupation', 'native-country','income-group','race', 'sex'])
f2 = f2.drop(columns=['workclass', 'education', 'marital-status', 'occupation', 'native-country','income-group','race', 'sex'])
```

```python
from recordlinkage.index import Full
from recordlinkage.index import Block
import recordlinkage

indexer = recordlinkage.Index()
indexer.block(left_on=['hours-per-week', 'relationship', 'capital-gain', 'capital-loss', 'age'],
              right_on=['hours-per-week', 'relationship', 'capital-gain', 'capital-loss', 'age'])
candidate_links = indexer.index(f1, f2)
```

```python
len(candidate_links)
```

```python
comp = recordlinkage.Compare()
comp.numeric('hours-per-week', 'hours-per-week', label='hours-per-week')
comp.numeric('capital-loss', 'capital-loss', label='capital-loss')
comp.numeric('capital-gain', 'capital-gain', label='capital-gain')
#comp.exact('age', 'age', label='age')
#comp.exact('relationship', 'relationship', label='relationship')
features = comp.compute(candidate_links, f1, f2)
```

```python
features.tail(100)
```

```python
unwrap = features.reset_index()
unwrap.head()
```

```python
counts = unwrap.groupby(['level_0'])['level_1'].count()
counts
```

```python
counts_df = counts.to_frame()
counts_df.head()
```

```python
completely_linked = counts_df[counts_df['level_1'] == 1]
len(completely_linked)
```
