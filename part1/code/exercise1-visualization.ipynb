```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
f1 = pd.read_csv("F1_final.csv")
f2 = pd.read_csv("F2_final.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plott(feature):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5))
    sns.histplot(kde=True, data=f1, x=feature, color='#ED174B', ax=ax1)
    sns.histplot(kde=True, data=f2, x=feature, color='#006BB7', ax=ax2)

    ax1.set_title("F1")
    ax2.set_title("F2")
    plt.show()
```

```python
numerics = ['capital-gain', 'capital-loss', 'hours-per-week']
for i in numerics:
    plott(i)
```

```python
def plotqi(feature):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5))
    sns.countplot(data=f1, x=feature, color='#ED174B', ax=ax1)
    sns.countplot(data=f2, x=feature, color='#006BB7', ax=ax2)

    ax1.set_title("F1")
    ax2.set_title("F2")
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)
    ax1.bar_label(ax1.containers[0])
    ax2.bar_label(ax2.containers[0])
    plt.show()
```

```python
QIs = ['sex', 'age', 'workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex']
for i in QIs:
    plotqi(i)
```
