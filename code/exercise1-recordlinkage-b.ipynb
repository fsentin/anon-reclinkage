```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv("adult_csv.csv")
names = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week","native-country", "income-group"]
df.columns = names
```

```python
for name in names:
    df.drop(df[df[name] == " ?"].index, inplace=True)
df.reset_index(inplace=True, drop=False)
df.drop("fnlwgt", axis=1, inplace=True)
df.drop("education-num", axis=1, inplace=True)
df.drop("index", axis=1, inplace=True)
```

```python
f1 = pd.read_csv("F1_final.csv")
```

```python
f1 = f1.drop(columns=['workclass', 'education', 'marital-status', 'occupation', 'native-country','income-group','race', 'sex', 'age'])
```

```python
def conditions(s):
    if s == 'Husband':
        return ' Male'
    elif s  == 'Wife':
        return ' Female'
    else:
        return ' Other'

f1['sex'] = f1['relationship'].apply(conditions)
```

```python
from recordlinkage.index import Full
from recordlinkage.index import Block
import recordlinkage

indexer = recordlinkage.Index()
indexer.block(left_on=['sex', 'hours-per-week', 'capital-gain', 'capital-loss'],
              right_on=['sex', 'hours-per-week', 'capital-gain', 'capital-loss'])
candidate_links = indexer.index(df, f1)
```

```python
len(candidate_links)
```

```python
comp = recordlinkage.Compare()
comp.numeric('hours-per-week', 'hours-per-week', label='hours-per-week')
comp.numeric('capital-loss', 'capital-loss', label='capital-loss')
comp.numeric('capital-gain', 'capital-gain', label='capital-gain')
features = comp.compute(candidate_links, df, f1)
```

```python
features.tail(100)
```

```python
unwrap = features.reset_index()
unwrap.head()
```

```python
counts = unwrap.groupby(['level_0'])['level_1'].count()
counts
```

```python
counts_df = counts.to_frame()
counts_df.head()
```

```python
completely_linked = counts_df[counts_df['level_1'] == 1]
len(completely_linked)
```
